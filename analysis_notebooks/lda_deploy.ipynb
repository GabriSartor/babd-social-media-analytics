```python
from database import  bulk_update_comments
```

```python
import bson
```

```python
import pandas as pd
```

```python
df = pd.read_csv("after_topic_prob.csv")
```

```python
bulk_size = 5000
total = 0
bulk = []
for index, row in df.iterrows():
    bulk.append([{ "_id": bson.objectid.ObjectId(row['Unnamed: 0']) }, 
    { "$set": 
        { 'Topic_1_levan' : row['Topic_1'], 
         'Topic_2_levan' : row['Topic_2'], 
         'Topic_3_levan' : row['Topic_3'], 
         'Topic_4_levan' : row['Topic_4'], 
         'Topic_5_levan' : row['Topic_5'], 
         'Topic_6_levan' : row['Topic_6'], 
         'Topic_7_levan' : row['Topic_7'], 
         'Topic_8_levan' : row['Topic_8'], 
         'Topic_9_levan' : row['Topic_9'], 
         'Topic_10_levan' : row['Topic_10']
        } 
    }])
    if len(bulk) > bulk_size:
        res = bulk_update_comments(bulk)
        total += len(bulk)
        print("Entry modified: {}".format(res.modified_count))
        print("Total progress: {}".format(total))
        bulk = []
```
